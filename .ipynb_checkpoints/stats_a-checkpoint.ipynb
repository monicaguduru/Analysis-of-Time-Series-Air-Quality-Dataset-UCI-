```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pylab as plt
from matplotlib import pyplot
import warnings
#from statsmodels.tsa.stattools import acf
#import pandas.plotting.autocorrelation_plot
from pandas.tools.plotting import autocorrelation_plot
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('AirQuality.csv ', header=0, na_values=[-200])
```

```python
print (data.head())
```

```python
print(data.shape)
```

```python
data = data.drop(columns=['Unnamed: 15','Unnamed: 16'])
```

```python
print (data.shape)
```

```python
data.mean()
all_cols = ['CO(GT)','PT08.S1(CO)','NMHC(GT)','C6H6(GT)','PT08.S2(NMHC)','NOx(GT)','PT08.S3(NOx)',
            'NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T','RH','AH']
```

```python
data = data.replace( -200 , np.nan )
```

```python
data = data.dropna(how="all")
```

```python
data = data.fillna(data.mean())
```

```python
print(data.head())
print(data.tail())
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
# Test for stationarity
# 1. plot and check
for i in all_cols:
    ts = data[i]
    plt.plot(ts)
    plt.show()
```

```python
for i in all_cols:
    ts = data['CO(GT)']
    autocorrelation_plot(np.log(ts))
    plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from numpy import log

for i in all_cols:
    ts= data[i]
    result = adfuller(log(ts.values))
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

```
